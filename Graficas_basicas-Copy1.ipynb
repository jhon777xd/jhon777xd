```python
#COMO GENERAR UNA GRAFICA LINEAL
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(1,5)
plt.plot(x, x*3.8, label="sulfuro")
plt.plot(x, x*4.9, label="oxido")
plt.plot(x, x/6.9, label="zinc") 

plt.grid(True, which="major", color="green", linestyle="-") 
plt.minorticks_on()
plt.grid(b=True, which="minor", color="#999999", linestyle="-", alpha=0.2)
plt.legend(loc="upper left");
```

```python
#COMO GENERAR UNA GRAFICA DE BARRRAS CON LOS ERRORES DETERMINADOS
import matplotlib.pyplot as plt
import numpy as np

data1= 10*np.random.rand(10)
data2= 10*np.random.rand(10)
data3= 10*np.random.rand(10)

e2 = 0.5*np.abs(np.random.randn(len(data1)))

locs = np.arange(1, len(data1)+1)
width=0.27

plt.bar(locs, data1, width=width)
plt.bar(locs+width, data2, yerr=e2, width=width, color="red")
plt.bar(locs+2*width, data3, width=width, color="green")

plt.xticks(locs+1.1*width, locs);
```

```python
#COMO GENERAR UNA GRAFICA POLAR
import matplotlib.pyplot as plt
import numpy as np

Beta = np.arange(0., 2., 1./360)*np.pi

plt.polar(2*Beta, Beta/3, color="skyblue")
plt.polar(Beta, np.cos(4*Beta), color="green")
plt.polar(Beta, [1.4]*len(Beta), color="red");
```

```python
#DOBLE EJE
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0., np.e, 0.001)
y1 = np.exp(-x)
y2 = np.log(x)

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(x, y1)
ax1.set_ylabel("Y valores para exp(-x)");
ax2 = ax1.twinx()
ax2.plot(x, y2, "r")
ax2.set_ylabel("Y valores para ln(x)");
ax1.set_xlabel("Mismo eje x")

#GENERAR ESTAS LINEAS ES OPCIONAL
plt.grid(True, which="major", color="green", linestyle="-") 
plt.minorticks_on()
plt.grid(b=True, which="minor", color="#999999", linestyle="-", alpha=0.2)
```

```python
#GRAFICA TRIDIMENSIONAL
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
x = np.arange(-4, 4, 0.25)
y = np.arange(-4, 4, 0.25)
x, y = np.meshgrid(x,y)
R = np.sqrt(x**2 + y**2)
z = np.sin(R)

ax.plot_surface(x, y, z, cmap="hot")
```
